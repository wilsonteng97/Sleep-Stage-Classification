```python
import pandas as pd
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
import os
import tensorflow as tf
```

```python
NUM_CLASS=5
BATCH_SIZE = 32
BETA = 1e-9
EPOCHS = 1000
LR = 0.01
NUM_HIDDEN = 20
SEED = 10

np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
def layer3():
    model = keras.Sequential()
    # hidden
    model.add(
        keras.layers.Dense(
            NUM_HIDDEN,
            activation='relu',
            kernel_regularizer=keras.regularizers.l2(BETA)
        )
    )
    # output
    model.add(

        keras.layers.Dense(
            NUM_CLASS,
            activation='softmax',
            kernel_regularizer=keras.regularizers.l2(BETA)
        )
    )
    return model
```

```python
train_input = np.genfromtxt('train.csv', delimiter= ',')
trainX = train_input[1:, 1:].astype(float)
train_Y =  train_input[1:,0].astype(int)
y_train = np.zeros((train_Y.shape[0], NUM_CLASS))
y_train[np.arange(train_Y.shape[0]), train_Y] = 1

test_input = np.genfromtxt('test.csv', delimiter= ',')
testX = test_input[1:, 1:].astype(float)
test_Y = test_input[1:,0].astype(int)
y_test = np.zeros((test_Y.shape[0], NUM_CLASS))
y_test[np.arange(test_Y.shape[0]), test_Y] = 1
```

```python
# y_train = tf.Variable(y_train)
# X_train = tf.Variable(X_train)
# y_test = tf.Variable(y_test)
# X_test = tf.Variable(X_test)
```

```python
model = layer3()

model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=LR),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(trainX, y_train, validation_data=(testX, y_test),
                                        epochs=EPOCHS,
                                        verbose = 2,
                                        batch_size=BATCH_SIZE,
                                        shuffle=True)
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv('basic.csv')
```

```python

from sklearn import preprocessing

x = data.values 
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data = pd.DataFrame(x_scaled)
```

```python
data.head()
```

```python
size = data.shape[0]
idx = np.arange(size)
np.random.shuffle(idx)
split = int(0.7 * size)
train_idx, test_idx = idx[:split], idx[split:]
data.iloc[train_idx].to_csv("train.csv", index=False)
data.iloc[test_idx].to_csv("test.csv", index=False)
```
